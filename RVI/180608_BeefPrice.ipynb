```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import warnings
import seaborn as sns

warnings.filterwarnings('ignore')
font_name = matplotlib.font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
krfont = {'family':font_name, 'size':10}
matplotlib.rc('font', **krfont)
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 불러오기
price = pd.read_csv('./data/Price_201803.csv', encoding='CP949')
price.drop(['조사구분코드','표준품목코드','조사가격품목코드','표준품종코드','조사가격품종코드','조사등급코드','표준단위코드','조사지역코드','표준시장코드','조사가격시장코드','표준단위명'], axis=1, inplace=True)
price.info()
price.head(3)
```
