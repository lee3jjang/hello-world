```python
import pandas as pd
#import numpy as np
import os
#import sys
#import matplotlib.pyplot as plt
#import matplotlib
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import random
import time
from datetime import datetime
```

```python
driver = webdriver.Chrome('chromedriver')
```

```python
car_domestic = 'http://www.encar.com/dc/dc_carsearchlist.do?carType=kor&searchType=model&TG.R=A#!'
car_foreign = 'http://www.encar.com/fc/fc_carsearchlist.do?carType=for&searchType=model&TG.R=B#!'
car_ev = 'http://www.encar.com/ev/ev_carsearchlist.do?carType=ev&searchType=model&TG.R=D#!'
page_num = 100   # 수집 페이지 수
delay = 1.5   # 딜레이 시간
```

```python
# 접속하기
driver.get(car_ev)

# [20개씩 보기] → [50개씩 보기]로 변환
viewer = Select(driver.find_element_by_css_selector('select#pagerow'))
viewer.select_by_value('50')
time.sleep(delay)

# 차종 카테고리 열기
driver.find_element_by_css_selector('div.schset.category').find_element_by_css_selector('h5').click()
time.sleep(delay)
```

```python
# 차종 선택
for category_num in range(12):
    if category_num != 0:
        driver.find_element_by_css_selector('input#category_{}'.format(category_num-1)).click()
        time.sleep(delay)
    driver.find_element_by_css_selector('input#category_{}'.format(category_num)).click()
    time.sleep(delay)
    car_category = driver.find_element_by_css_selector('input#category_{}'.format(category_num)).find_element_by_xpath('./following-sibling::label').text
    
    # 전기ㆍ친환경차 종류 선택
    for ev_type_num in range(4):
        if ev_type_num != 0:
            driver.find_element_by_css_selector('input#evType_{}'.format(ev_type_num-1)).click()
            time.sleep(delay)
        driver.find_element_by_css_selector('input#evType_{}'.format(ev_type_num)).click()
        time.sleep(delay)
        ev_type = driver.find_element_by_css_selector('input#evType_{}'.format(ev_type_num)).find_element_by_xpath('./following-sibling::label').text
        ################
        #crawling(car_category, ev_type)
        ################
    driver.find_element_by_css_selector('input#evType_{}'.format(ev_type_num)).click()
    time.sleep(delay)
```

```python
# 데이터 수집하기
#print('----- 수집 시작 -----')
result = []

page = 1
#for page in range(1, page_num+1):
# 다음 페이지로 이동
#driver.find_element_by_css_selector('div#pagination').find_element_by_xpath('//a[@data-page="{}"]'.format(page)).click()
# 기다리기
#WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'sr_normal')))
# Table 집어오기
#temp = pd.read_html(driver.find_element_by_css_selector('tbody#sr_normal').find_element_by_xpath('..').get_attribute('outerHTML'))[0]

car_category = '경차'
ev_type = '하이브리드'
def get_data_batch():
    result = []
    rows = driver.find_element_by_css_selector('tbody#sr_normal').find_elements_by_tag_name('tr')
    for row in rows:
        dignose = row.find_element_by_css_selector('td.img > div> a > span.blt').text

        name1 = row.find_element_by_css_selector('td.inf > a > span.cls > strong').text
        name2 = row.find_element_by_css_selector('td.inf > a > span.cls > em').text
        name3 = row.find_element_by_css_selector('td.inf > a > span.dtl > strong').text
        name4 = row.find_element_by_css_selector('td.inf > a > span.dtl > em').text

        yer = row.find_element_by_css_selector('td.inf > span.detail > span.yer').text
        km = row.find_element_by_css_selector('td.inf > span.detail > span.km').text
        fue = row.find_element_by_css_selector('td.inf > span.detail > span.fue').text
        loc = row.find_element_by_css_selector('td.inf > span.detail > span.loc').text
        try:
            ins = row.find_element_by_css_selector('td.inf > span.detail > span.ins').text
        except:
            ins = ''
        try:
            ass = row.find_element_by_css_selector('td.inf > span.detail > span.ass').text
        except:
            ass = ''

        hotmarks = row.find_element_by_css_selector('td.inf > span.hotMark').text
        prc_hs = row.find_element_by_css_selector('td.prc_hs').text
        svc = row.find_element_by_css_selector('td.svc').text

        result.append([dignose, name1, name2, name3, name4, yer, km, fue, loc, ins, ass, prc_hs, svc, hotmarks])
    df = pd.DataFrame(result)
    df.columns = ['진단여부', '이름1', '이름2', '이름3', '이름4', '연식', '주행거리', '연료', '지역', '성능기록', '엔카진단', '가격', 'SVC', '핫마크']
    df.insert(0, '종류2', ev_type)
    df.insert(0, '종류1', car_category)
    return df

df = get_data_batch()
#row.click()
#result.append(temp)
#time.sleep(random.random()*2+1)
#print('----- 수집 완료 -----')
```
