```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.linspace(-5, 5, 1000)
y = sigmoid(x)
plt.plot(x, y)
plt.plot([0, 0], [1, 0], ':')
plt.title('Sigmoid Function')
plt.show()
```
