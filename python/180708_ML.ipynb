```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigmoid = lambda x: 1/(1+np.exp(-x))

x = np.arange(-10,10,0.2)
sig = sigmoid(x)
plt.plot(x, sig)
plt.show()
```

```python
odds = lambda p: p/(1-p)
p = np.arange(0,1,0.01)
od = odds(p)
plt.plot(p, od)
plt.show()
```

```python
log_odds = lambda p: np.log(p/(1-p))
p = np.arange(0.005, 1, 0.005)
log_od = log_odds(p)
plt.plot(p, log_od)
plt.yticks([-4,0,4])
plt.grid(True, axis='y')
plt.show()
```

```python
# read csv
data = []
with open('./data/logiReg.csv', 'r') as f:
    reader = csv.reader(f, quoting=csv.QUOTE_NONNUMERIC)
    for row in reader:
        data.append(row)
data = np.array(data)
```

```python
scale = 50.0
a = 0.5

paid = data[np.where(data[:,2] == 1)]
unpaid = data[np.where(data[:,2] == 0)]
x, y = paid[:,0], paid[:,1]
plt.scatter(x,y, s=scale, label='paid', alpha=a, edgecolors='none')
x, y = unpaid[:,0], unpaid[:,1]
plt.scatter(x,y, s=scale, label='unpaid', alpha=a, edgecolors='none')

plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(data[:,:2])
X = scaler.transform(data[:,:2])
X = np.insert(X,0,1,axis=1)
y = data[:,2]
theta = np.ones(3)
```

```python
f = lambda x,theta: sigmoid(np.dot(x, theta))
loss = lambda theta,x,y: -np.mean(y*np.log(f(x, theta)) + (1-y)*np.log(1-f(x,theta)))
```

```python
from functools import partial
from scipy.optimize import minimize
loss_partial = partial(loss, x=X, y=y)
res = minimize(loss_partial, x0=theta, method='nelder-mead')
fitted =  res.x
y_pred = np.where(f(X,fitted) > 0.5,1,0)
y_true = y.astype(int)
fitted
```

```python
from sklearn.metrics import confusion_matrix
tn, tp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
confusion_matrix(y_true, y_pred)
```
