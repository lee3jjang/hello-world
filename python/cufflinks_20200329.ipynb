```python
import numpy as np
import pandas as pd
import cufflinks as cf
cf.go_offline(connected=True)
```

```python
df = cf.datagen.lines()
df.iplot(kind='line')
```

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['A', 'B', 'C', 'D'])
df.iplot(kind='bar')
```

```python
df.iplot(kind='bar', barmode='stack')
```

```python
df['A'].iplot(kind='bar')
```

```python
df.iplot(kind='barh', barmode='stack')
```

```python
df.iplot(kind='area')
```

```python
df.iplot(kind='area', fill=True)
```

```python
df.iplot(fill=True)
```

```python
df.iplot(kind='bar', theme='space')
```

```python
cf.set_config_file(theme='ggplot')
df.iplot()
```

```python
layout = {
    'font': {
        'family': 'consolas',
        'size': 20,
        'color': 'blue'
    },
    'title': 'Title'
}

df.iplot(kind='bar', layout=layout)
```
