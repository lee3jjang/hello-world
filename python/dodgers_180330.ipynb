```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))

dodgers = pd.read_csv('./data/dodgers.csv')
dodgers.head()

sns.set(style='darkgrid')
plt.scatter(dodgers.temp, dodgers.attend/1000)
plt.title('Temperature by Attendance')
plt.xlabel('Temperature')
plt.ylabel('Attendance(thousands)')
```

```python
month_to_num = lambda x:{'APR':4,'MAY':5,'JUN':6,'JUL':7,'AUG':8,'SEP':9}.get(x,10)
dodgers['ordered_month'] = dodgers['month'].apply(month_to_num)
```

```python
fig, axes = plt.subplots(1,2, figsize=(16,6), sharex=True)

sns.set_style('whitegrid')
fig = sns.boxplot(y="attend", x="ordered_month", data=dodgers, ax=axes[0])
sns.violinplot(y="attend", x="ordered_month", data=dodgers, ax=axes[1])

for ax in axes:
    ax.set(xlabel='Month', ylabel='Attendance')
```

```python
week_to_num = lambda x:{'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6}.get(x,7)
dodgers['ordered_day_of_week'] = dodgers['day_of_week'].apply(week_to_num)
```
