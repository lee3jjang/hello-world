```python
import numpy as np
import pandas as pd
from dfply import *
```

```python
np.random.seed(20190428)

n = 50
특징 = np.random.randint(3, size=n)
값 = np.random.normal(3, 1.5, size=n)

df = pd.DataFrame({'특징': 특징, '값': 값})
```

```python
# pct_change
res = list(df['값'].pct_change()[1:])

res2 = []
for i in range(len(df)-1):
    res2.append(df['값'].iloc[i+1]/df['값'].iloc[i]-1)
    
# 값 검증
print(np.sum((np.array(res)-np.array(res2))**2))
```

```python
# diff
res3 = df['값'].diff()[1:]

res4 = []
for i in range(len(df)-1):
    res4.append(df['값'].iloc[i+1]-df['값'].iloc[i])
    
# 값 검증
print(np.sum((np.array(res3)-np.array(res4))**2))
```

```python
# cumsum
res = list(df['값'].cumsum(axis=0))

res2 = []
for i in range(len(df)):
    res2.append(df['값'].iloc[:(i+1)].sum())
    
# 값 검증
print(np.sum((np.array(res) - np.array(res2))**2))
```

```python
flight_data = pd.read_csv('nycflights13.csv')
```

```python
flight_data.head()
```

```python
df = flight_data >> select(X.origin, X.dest, X.hour)
df.head()
```

```python
df = flight_data >> drop(X.year, X.month, X.day)
df.head()
```

```python
df = flight_data >> select(~X.hour, ~X.minute)
df.head()
```

```python
df = flight_data >> mask(X.month==1, X.day==1, X.origin=='JFK', X.hour>10)
df.head()
```

```python
df = flight_data >> arrange(X.distance, X.hour, ascending=[True, False])
df.head()
```

```python
df = flight_data >> mutate(
    new_distance = X.distance / 1000,
    carrier_origin = X.carrier + X.origin
)
df.head()
```

```python
df = flight_data >> group_by(X.origin) >> summarize(mean_distance = X.distance.mean())
df.head()
```

```python
df = flight_data >> group_by(X.origin) >> summarize(count=X.origin.count()) >> right_join(flight_data, by='origin')
df.head()
```

```python
df = flight_data >> group_by(X.origin) >> summarize(count=X.origin.count()) >> right_join(flight_data, by='origin')
df.head()
```

```python
df = flight_data >> group_by(X.origin) >> mutate(flight_lag=lag(X.flight)) >> select(X.origin, X.flight, X.flight_lag)
df.head()
```

```python
df = flight_data >> group_by(X.origin) >> mutate(flight_lead=lead(X.flight)) >> select(X.origin, X.flight, X.flight_lead)
df.head()
```
