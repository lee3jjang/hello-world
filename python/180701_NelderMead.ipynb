```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rosen(x, y):
    return x**2 - 4*x + y**2 - y - x*y
```

```python
v = np.array([(0,0),(1.2,0),(0,0.8)])
b, g, w = v[np.array(list(map(lambda x:rosen(x[0], x[1]), v))).argsort()]

pts = [b,g,w]
names = ['B','G','W']
n = len(pts)
for i in range(n):
    plt.scatter(pts[i][0], pts[i][1])
    plt.text(pts[i][0]+0.2, pts[i][1]+0.2, names[i], fontsize=12)
    plt.text(pts[i][0]+0.3, pts[i][1]+0.2, ' {:.2f}'.format(rosen(pts[i][0], pts[i][1])), fontsize=12)

plt.title('Initial Triangle')
plt.grid(True)
plt.xlim(-1.5,2.5)
plt.ylim(-1.5,2.5)
plt.show()
```

```python
m = (b+g)/2

pts = [b,g,w,m]
names = ['B','G','W','M']
n = len(pts)
for i in range(n):
    plt.scatter(pts[i][0], pts[i][1])
    plt.text(pts[i][0]+0.2, pts[i][1]+0.2, names[i], fontsize=12)
    plt.text(pts[i][0]+0.3, pts[i][1]+0.2, ' {:.2f}'.format(rosen(pts[i][0], pts[i][1])), fontsize=12)

plt.title('Midpoint')
plt.grid(True)
plt.xlim(-1.5,2.5)
plt.ylim(-1.5,2.5)
plt.show()
```

```python
r = 2*m-w

pts = [b,g,w,m,r]
names = ['B','G','W','M','R','E']
n = len(pts)
for i in range(n):
    plt.scatter(pts[i][0], pts[i][1])
    plt.text(pts[i][0]+0.2, pts[i][1]+0.2, names[i], fontsize=12)
    plt.text(pts[i][0]+0.3, pts[i][1]+0.2, ' {:.2f}'.format(rosen(pts[i][0], pts[i][1])), fontsize=12)

plt.title('Reflection')
plt.grid(True)
plt.xlim(-1.5,2.5)
plt.ylim(-1.5,2.5)
plt.show()
```

```python
e = 2*r-m

pts = [b,g,w,m,r,e]
names = ['B','G','W','M','R','E']
n = len(pts)
for i in range(n):
    plt.scatter(pts[i][0], pts[i][1])
    plt.text(pts[i][0]+0.2, pts[i][1]+0.2, names[i], fontsize=12)
    plt.text(pts[i][0]+0.3, pts[i][1]+0.2, ' {:.2f}'.format(rosen(pts[i][0], pts[i][1])), fontsize=12)

plt.title('Expansion')
plt.grid(True)
plt.xlim(-1.5,2.5)
plt.ylim(-1.5,2.5)
plt.show()
```

```python
c = [(w+m)/2, (m+r)/2]
c = c[np.argmin(np.array([rosen(c[0][0], c[0][1]), rosen(c[1][0], c[1][1])]))]

pts = [b,g,w,m,r,e,c]
names = ['B','G','W','M','R','E','C']
n = len(pts)
for i in range(n):
    plt.scatter(pts[i][0], pts[i][1])
    plt.text(pts[i][0]+0.2, pts[i][1]+0.2, names[i], fontsize=12)
    plt.text(pts[i][0]+0.3, pts[i][1]+0.2, ' {:.2f}'.format(rosen(pts[i][0], pts[i][1])), fontsize=12)

plt.title('Contraction')
plt.grid(True)
plt.xlim(-1.5,2.5)
plt.ylim(-1.5,2.5)
plt.show()
```

```python
s = (w+b)/2

pts = [b,g,w,m,r,e,c,s]
names = ['B','G','W','M','R','E','C','S']
n = len(pts)
for i in range(n):
    plt.scatter(pts[i][0], pts[i][1])
    plt.text(pts[i][0]+0.2, pts[i][1]+0.2, names[i], fontsize=12)
    plt.text(pts[i][0]+0.3, pts[i][1]+0.2, ' {:.2f}'.format(rosen(pts[i][0], pts[i][1])), fontsize=12)

plt.title('Contraction')
plt.grid(True)
plt.xlim(-1.5,2.5)
plt.ylim(-1.5,2.5)
plt.show()
```
