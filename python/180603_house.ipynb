```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import warnings
warnings.filterwarnings('ignore')
font_name = matplotlib.font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
krfont = {'family':font_name, 'weight':'bold', 'size':10}
matplotlib.rc('font', **krfont)
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
train = pd.read_csv('../data/house/train.csv')
train.info()
```

```python
train.head(2)
```

```python
sns.distplot(np.log(train['SalePrice']))
plt.show()
```

```python
sns.set_style('whitegrid')
plt.scatter(train['LotArea'], train['SalePrice'], s=10, c='skyblue', alpha=0.7, marker='s')
plt.xscale('log')
plt.show()
```

```python
sns.boxplot(x='YrSold', y='SalePrice', data=train)
```

```python
groups = train.groupby('YrSold')['SalePrice']

for k, v in groups:
    v.hist(alpha=0.65, label=str(k), bins=50)

plt.legend()
plt.show()
```

```python
groups = train.groupby('YrSold')['SalePrice']

for k, v in groups:
    sns.kdeplot(np.log(v), label=str(k))

plt.legend()
plt.show()
```

```python
train.groupby('YrSold').hist(column='SalePrice', bins=50)
plt.show()
```
