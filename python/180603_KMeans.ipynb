#### 1. K-Means

```python
sns.set_style('white')
X, y = make_blobs(n_samples=150,
                  n_features=2,
                  centers=3,
                  cluster_std=0.5,
                  shuffle=True,
                  random_state=0)

plt.scatter(X[:,0], X[:,1], c=y, marker='o', s=50)
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=0)
clsf = kmeans.fit_predict(X)
plt.scatter(X[clsf==0,0], X[clsf==0,1], c='lightgreen', marker='s', s=50, label='Cluster 1')
plt.scatter(X[clsf==1,0], X[clsf==1,1], c='orange', marker='o', s=50, label='Cluster 2')
plt.scatter(X[clsf==2,0], X[clsf==2,1], c='lightblue', marker='v', s=50, label='Cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='red', marker='*', s=50, label='Center')

plt.legend()
plt.grid(True)
plt.show()
```

#### 2. DBSCAN : Density-based Spatial Clustering of Applications with Noise

```python
X, y = make_moons(n_samples=200,
                  noise=0.05,
                  random_state=0)

plt.scatter(X[:,0], X[:,1], c=y)
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
clsf = dbscan.fit_predict(X)
plt.scatter(X[clsf==0,0], X[clsf==0,1], c='lightgreen', marker='s', s=50, label='Cluster 1')
plt.scatter(X[clsf==1,0], X[clsf==1,1], c='orange', marker='o', s=50, label='Cluster 2')

plt.grid(True)
plt.show()
```
