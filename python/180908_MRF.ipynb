```python
eta = 2.0
xi = 1.5
```

```python
def delE(img, i, j):
    left = right = up = down = 0
    if j != 0:
        left = img[i,j-1]
    if j != img.shape[1]-1:
        right = img[i,j+1]
    if i != 0:
        up = img[i-1,j]
    if i != img.shape[0]-1:
        down = img[i+1,j]
    E = 2*img[i,j]*(left+right+up+down)
    return E
```

```python
row = 100
col = 100
img3 = img2[:row,:col].copy()

i = 0
flag = True
while(flag == True):
    i += 1
    flag = False
    tmp = img3.copy()
    for i in range(row):
        for j in range(col):
            cr = delE(tmp, i, j)
            if cr < -eta:
                flag =  True
                img3[i,j] = -img3[i,j]
print(i)                
```
