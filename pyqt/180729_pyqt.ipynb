```python
import sys
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import QApplication, QWidget, QLCDNumber, QSlider, QVBoxLayout
class Ex(QWidget):
    
    def __init__(self):
        super().__init__()
        self.initUI()
        return None
    
    def initUI(self):
        
        # 숫자 나오는 LCD 숫자 객체
        lcd = QLCDNumber(self)
        
        # 가로방향의 슬라이더 객체
        sld = QSlider(Qt.Horizontal, self)
        
        #VBox 레이아웃
        vbox = QVBoxLayout()
        
        #LDC와 슬라이더를 VBox에 추가
        vbox.addWidget(lcd)
        vbox.addWidget(sld)
        
        # VBox 레이아웃 설정?
        self.setLayout(vbox)
        # 슬라이더 숫자가 바뀌면 lcd의 display와 연결되게 함
        sld.valueChanged.connect(lcd.display)
        
        self.setGeometry(300,300,250,150)
        self.setWindowTitle('시그널이랑 슬롯')
        self.show()
        return None
    
    # keyPressEvent라는 이름을 무조건 써야함
    def keyPressEvent(self, e):
        if e.key() == Qt.Key_Escape:
            self.close()
    
if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = Ex()
    sys.exit(app.exec_())
```

```python
import sys
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import QApplication, QWidget, QGridLayout, QLabel
class Ex(QWidget):
    
    def __init__(self):
        super().__init__()
        self.initUI()
        return None
    
    def initUI(self):
        # 레이아웃은 self로 들고 있지 않고, 객체들 받아서 setLayout으로 화면에 꽂아주는 듯?
        grid = QGridLayout()
        x = 0
        y = 0
        self.text = "X: {}, Y: {}".format(x,y)
        
        # QLabel 객체인데 arg1는 텍스트, args2는 self
        self.label = QLabel(self.text, self)
        
        # 이게 있어야 그리드 안에 라벨이 들어가는 듯
        grid.addWidget(self.label, 0, 0, Qt.AlignTop)
        
        # 이게 있어야 레이아웃이 잘 나오는 듯
        self.setLayout(grid)
        # 이게 있어야 자동으로 마우스를 따라가님
        self.setMouseTracking(True)
        self.setGeometry(300,300,250,150)
        self.setWindowTitle('시그널이랑 슬롯')
        self.show()
        return None
        
    
    # keyPressEvent라는 이름을 무조건 써야함
    def keyPressEvent(self, e):
        if e.key() == Qt.Key_Escape:
            self.close()

    # mouseMoveEvent라는 이름을 무조건 써야함
    def mouseMoveEvent(self, e):
        x = e.x()
        y = e.y()
        text = "X: {}, Y: {}".format(x,y)
        self.label.setText(text)
    
if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = Ex()
    sys.exit(app.exec_())
```

```python
import sys
from PyQt5.QtCore import Qt
from PyQt5.QtWidgets import QApplication, QMainWindow, QPushButton
class Ex(QMainWindow):
    
    def __init__(self):
        super().__init__()
        self.initUI()
        return None
    
    def initUI(self):
        btn1 = QPushButton('Button 1', self)
        btn1.move(30, 50)
        
        btn2 = QPushButton('Button 2', self)
        btn2.move(150, 50)
        
        btn1.clicked.connect(self.butt)
        btn2.clicked.connect(self.butt)
        
        self.statusBar()       
        self.setGeometry(300,300,290,150)
        self.setWindowTitle('시그널이랑 슬롯')
        self.show()
        return None
        
    def butt(self):
        # QMainWindow라는 객체에는 sender라는 메소드가 있음
        # 이거는 어떤 객체를 리턴하는데 여기에 text() 안에 버튼 이름이 있음
        sender = self.sender()
        self.statusBar().showMessage(sender.text()  + '가 눌렸습니다.')
        return None
    
if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = Ex()
    sys.exit(app.exec_())
```

```python
import sys
from PyQt5.QtCore import Qt, pyqtSignal, QObject
from PyQt5.QtWidgets import QApplication, QMainWindow

class Comm(QObject):
    closeApp = pyqtSignal()


class Ex(QMainWindow):
    
    def __init__(self):
        super().__init__()
        self.initUI()
        return None
    
    def initUI(self):
        # Comm 객체 생성
        self.c = Comm()
        # Comm 객체 안에 closeApp(pyqtSignal() 객체)는 close랑 연결
        self.c.closeApp.connect(self.close)
        
        self.setGeometry(300,300,290,150)
        self.setWindowTitle('시그널이랑 슬롯')
        self.show()
        return None
        
    # mousePressEvent라는 이름을 무조건 써야함
    def mousePressEvent(self, e):
        # 마우스를 입력하면 pyqtSignal() 객체가 가진 emit 메소드를 실행
        self.c.closeApp.emit()
        return None
        
if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = Ex()
    sys.exit(app.exec_())
```
