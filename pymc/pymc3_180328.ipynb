```python
import pymc3_180328 as md
import pymc3 as pm
import matplotlib.pyplot as plt
import seaborn as sns

with md.bioassay_model:
    trace = pm.sample(1000, njobs=2)
    pm.forestplot(trace, varnames=['alpha', 'beta'])
```

```python
sns.set()
plt.hist(trace['alpha'], bins=100)
plt.hist(trace['beta'] , bins=100)
plt.show()
```

```python
sns.set_style('whitegrid')
current_palette = sns.color_palette()
plt.hist(trace['alpha'], bins=100, color=current_palette[4], alpha=0.8)
plt.hist(trace['beta'] , bins=100 , color=current_palette[5], alpha=0.8)
plt.show()
```

```python
sns.palplot()
```

```python
sns.kdeplot(trace['alpha'])
sns.distplot(trace['alpha'], bins=100)
plt.show()
```

```python
sns.distplot(trace['alpha'], kde=True, rug=False, bins=100)
plt.show()
```
