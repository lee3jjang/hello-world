```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def golden_section_search(fn, x1, x2, eps=1e-5):
    
    phi = (1+np.sqrt(5))/2
    x3 = x2 - (x2-x1)/phi
    x4 = x1 + (x2-x1)/phi
    x_trace = [(x1, x2)]
    while abs(x3-x4) > eps:
        if fn(x3) < fn(x4):
            x2 = x4
            x4 = x3
            x3 = x2 - (x2-x1)/phi
        else:
            x1 = x3
            x3 = x4
            x4 = x1 + (x2-x1)/phi
        x_trace.append([x1, x2])
    
    return (x1+x2)/2, x_trace
```

```python
test_func = lambda x: 2*np.sin(x)+3*np.cos(1.3*x)+0.01*(x-3)**3
```

```python
z, x_trace = golden_section_search(test_func, 0, 4*np.pi)
x = np.linspace(0, 4*np.pi, 100)
y = test_func(x)

plt.plot(x, y)
for x1, x2 in x_trace:
    plt.scatter([x1, x2], [test_func(x1), test_func(x2)])
#plt.scatter(z, test_func(z))
plt.title('Test Function')
plt.show()
```
