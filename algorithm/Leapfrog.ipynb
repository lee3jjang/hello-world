```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def grad_U(x):
    return 2*x

def Leapfrog(x,p,e):
    p = p - (e/2)*grad_U(x)
    x = x + e*p
    p = p - (e/2)*grad_U(x)
    return x, p

def Leapfrog2(x,p,e,L):
    result = [[x,p]]
    for _ in range(L):
        x,p = Leapfrog(x,p,e)
        result.append([x,p])
    result = pd.DataFrame(result, columns=['x','p'])
    return result
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,12))
for i in np.linspace(0,10,30):
    df = Leapfrog2(i,0,0.01,650)
    plt.plot(df.x,df.p)
plt.xlim([-15,15])
plt.ylim([-15,15])
plt.show()
```
