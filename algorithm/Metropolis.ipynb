```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def p(x):
    return np.exp(-x**2/2)

x0 = np.random.normal(0,1)
n = 100000
result = [x0]


for _ in range(n):
    xp = np.random.normal(x0,1)
    alpha = min(1,p(xp)/p(x0))
    u = np.random.random()
    if u < alpha:
        x0 = xp
    else:
        x0 = x0
    result.append(x0)
```

```python
sns.set_style('whitegrid')
sns.distplot(result, bins=100, label='Metropolis')
sns.distplot(np.random.normal(0,1,100000), bins=100, label='Actual')
plt.legend(loc='best')
plt.show()
print('mu : {}\nsigma = {}'.format(np.mean(result), np.std(result)))
```
