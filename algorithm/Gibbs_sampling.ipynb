```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

rho = 0.8
mu0 = 0
mu1 = 0

x = np.random.normal(0,1,size=2)
result = [[x[0],x[1]]]
n = 1000

for _ in range(n):
    x0 = np.random.normal(mu0+rho*(x[1]-mu1),np.sqrt(1-rho**2))
    x1 = np.random.normal(mu1+rho*(x0-mu0),np.sqrt(1-rho**2))
    x = [x0,x1]
    result.append(x)
    
result = pd.DataFrame(result, columns=['x','y'])
```

```python
sns.axes_style('white')
g = sns.jointplot(x='x',y='y', data=result, kind='kde')
g.plot_joint(plt.scatter, c='r', s=10, linewidth=1, marker='+')
plt.show()
```

```python
sns.jointplot(x='x',y='y', data=result)
plt.show()
```

```python
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(result.x,result.y, cmap=cmap, shade=True, n_levels=60)
plt.show()
```
