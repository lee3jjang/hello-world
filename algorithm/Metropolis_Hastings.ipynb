```python
import numpy as np
from scipy.special import gamma
import matplotlib.pyplot as plt
import seaborn as sns

def p(A):
    y = 1.5
    B = 1
    return B**A*np.exp(-B*y)*y**(A-1)/gamma(A) * np.sin(np.pi*A)**2

x0 = -5*np.log(1-np.random.random())
n = 1000000
result = [x0]


for _ in range(n):
    xp = -5*np.log(1-np.random.random())
    c = np.exp(-x0/5)/np.exp(-xp/5)
    alpha = min(1,p(xp)/p(x0)*c)
    u = np.random.random()
    if u < alpha:
        x0 = xp
    else:
        x0 = x0
    result.append(x0)
```

```python
sns.set_style('whitegrid')
sns.distplot(result, bins=1000, label='Metropolis')
plt.show()
```
